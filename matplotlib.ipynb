## Basic Plotting


Create a simple plot of a lineal function.


```python
f = lambda x: x
x = np.linspace(-10, 10, 10)
y = f(x)

plt.plot(x, y, marker="o", linestyle="--", label="f(x) = x + 1")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("$f(x) = x + 1$")
plt.legend(fontsize=12, loc="upper left", frameon=False)
plt.show()
```

Create two subplots in the same figure.


```python
x_1 = np.linspace(0, 2 * np.pi, 100)
y_1 = np.sin(x_1)

x_2 = np.linspace(0, 2 * np.pi, 100)
y_2 = np.cos(x_2)

figure, axes = plt.subplots(1, 2, figsize=(5, 5), constrained_layout=True)

axes[0].plot(x_1, y_1, label="sin(x)", color="red")
figure.get_axes()[0].set_xlabel("x")
figure.get_axes()[0].set_ylabel("sin(x)")
figure.get_axes()[0].set_title("$sin(x)$")
axes[0].legend()

axes[1].plot(x_2, y_2, label="cos(x)", color="blue")
figure.get_axes()[1].set_xlabel("x")
figure.get_axes()[1].set_ylabel("cos(x)")
figure.get_axes()[1].set_title("$cos(x)$")
axes[1].legend()

plt.show()
```

Show randoms points in a scatter plot.


```python
x = np.random.randn(100)
y = np.random.randn(100)

plt.scatter(x, y, marker="^", color="green", label="Data points")
plt.scatter(y, x)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Scatter plot")
plt.legend()
plt.show()
```

Create a bar plot with each category of a categorical variable.


```python
x = ["A", "B", "C", "D", "E"]
y = np.random.random(5)

plt.bar(x, y, color="orange", label="Data points")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Bar plot")
plt.legend()
plt.show()
```

Create a histogram of a numeric variable.


```python
x = np.random.normal(0, 1, 1000)

# Obtenemos la frecuencia de los datos
y = np.histogram(x, bins=30)

plt.hist(x, bins=30, color="purple", label="Data points", edgecolor="black")
plt.plot(y[1][:-1], y[0], color="red", label="Frequency")
plt.xlabel("x")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.legend()
plt.show()
```

Create a pie chart with the frequency of each category of a categorical variable.


```python
x = ["A", "B", "C", "D", "E"]
y = [10, 20, 30, 40, 50]

plt.pie(y, labels=x, autopct="%1.1f%%", startangle=90, shadow=True)
plt.title("Pie chart")
plt.show()
```
