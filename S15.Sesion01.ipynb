## Regresión Lineal Simple


**Análisis de varianza**

Consiste en descomponer la variabilidad de la variable dependiente $Y$ en variabilidad explicada por el modelo (Regresión) más variabilidad no explicada (error). Bajo la hipótesis de que existe una relación lineal entre la variable respuesta y regresora, se requiere realizar el siguiente contraste de hipótesis.

$$
Y = \beta_0 + \beta_1 X
$$

$H_0: \beta_1 = 0 \quad$ (No hay relación lineal entre $X$ e $Y$) (El modelo **no** es significativo)  
$H_1: \beta_1 \neq 0 \quad$ (Hay relación lineal entre $X$ e $Y$) (El modelo **si** es significativo)

Pasos:

1. Planteo de Hipótesis:
    - $H_0: \beta_1 = 0$ (El modelo **no** es significativo)
    - $H_1: \beta_1 \neq 0$ (El modelo **si** es significativo)
2. Cálculo del Estadística de Contraste (construir cuadro Anova)  
   $F = \frac{\frac{SCR}{1}}{\frac{SCE}{n-2}} = \frac{CMR}{CME}$
3. Gráfica de la Región Crítica  
   Determinar el valor crítico $F_{(1 - \alpha, \ k, \ n - k - 1)}$
4. Decisión  
   Si $F_{\text{calculado}} > F_{\text{crítico}}$, rechazo $H_0$, por tanto si existe regresión lineal entre con al menos $x_1$, $x_2$ y $y$  
   Si $F_{\text{calculado}} < F_{\text{crítico}}$, no rechazo $H_0$

**Residuos**

En un modelo de regresión, los residuos o errores son las diferencias entre los verdaderos valores que toma la variable dependiente y los valores estimados. Se calcula por tanto mediante:

$$
\hat{\epsilon}_i = y_i - \hat{y}_i
$$

**Suma cuadrados Total (SCT)**

$$
SCT = \sum_{i=1}^{n} (y_i - \bar{y})^2
$$
$$
gl = n - 1
$$

Permite medir la variabilidad total de la variable dependiente $Y$.

**Suma cuadrados explicado Regresión (SCR)**

$$
SCR = \sum_{i=1}^{n} (\hat{y}_i - \bar{y})^2
$$
$$
gl = 1
$$

Acumula la parte de la variabilidad de la variable dependiente que consigue explicar el modelo.

**Suma de cuadrados de los residuos o errores (SCE)**

$$
SCE = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$
$$
gl = n - 2
$$

Acumula la variabilidad de la variable dependiente que no consigue explicar con el modelo.

$$
SCT = SCR + SCE
$$

* $y_i$: i-ésimo valor de $Y$
* $\hat{y}_i$: i-ésimo valor de $\hat{Y}$

#### Ejercicio


Se cuenta con información de una muestra de $10$ fondos de inversión, respecto a su rentabilidad porcentual para un periodo de doce meses y el activo total (en millones $). Los datos se presentan en la siguiente tabla.

Pruebe con un nivel de significacia de $5 \%$, si existe la relación lineal entre el activo total y la rentabilidad (Anva).


```python
# Datos
x = [22, 18, 14, 8, 26, 10, 16, 19, 20, 7]
y = [29.3, 17.6, 16, 9.5, 22.6, 12.4, 18.5, 21.4, 18.2, 12.5]
n = len(y)

# Regresión lineal
b_1 = (n * sum([x[i] * y[i] for i in range(n)]) - sum(x) * sum(y)) / (
    n * sum([x[i] ** 2 for i in range(n)]) - sum(x) ** 2
)
b_0 = (sum(y) - b_1 * sum(x)) / n
regresion = lambda x: b_0 + b_1 * x

# Suma de cuadrados
y_med = round(sum(y) / n, 2)
datos = [
    [
        x[i],
        y[i],
        round(regresion(x[i]), 2),
        round((regresion(x[i]) - y_med) ** 2, 2),
        round((y[i] - regresion(x[i])) ** 2, 2),
    ]
    for i in range(len(x))
]
SCR = sum([dato[3] for dato in datos])
SCE = sum([dato[4] for dato in datos])
SCT = SCR + SCE

# Calculo del estadístico de prueba F_c
F_c = ((SCR / 1) / (SCE / (n - 2))).__round__(2)
alpha = 0.05
F_tab = st.f.ppf(1 - alpha, 1, n - 2).round(3)

datos.append(
    [
        "",
        rf"\bar{{Y}} = {y_med}",
        "",
        rf"\sum{{SCR}} = {{{SCR:.2f}}}",
        rf"\sum{{SCE}} = {{{SCE:.2f}}}",
    ]
)

tabla = tb.tabulate(
    [
        [f"${dato[i]} $" for i in range(len(dato))] for dato in datos
    ],
    headers=[
        r"$X$",
        r"$Y_i$",
        rf"$\hat{{Y}}_i = {b_0:.2f} {'+' if b_1 >= 0 else ''} {b_1:.2f} X_i$",
        r"$SRC = \sum{(\hat{y}_i - \bar{y})^2}$",
        r"$SCE = \sum{(y_i - \hat{y}_i)^2}$",
    ],
    tablefmt="github",
)

# Resultado
display(
    di.Markdown(
        tabla +
        "\n\n"
        "$$\n"
        "SCT = SCR + SCE"
        "\n$$\n\n"
        "$$\n"
        rf"SCT = {SCT:.2f} = {SCR:.2f} + {SCE:.2f}"
        "\n$$\n\n"
        "Paso 1:  \n"
        "$H_0: \\beta_1 = 0$ (No hay relación lineal entre $X$ y $Y$)  \n"
        "$H_1: \\beta_1 \\neq 0$ (Si hay relación lineal entre $X$ y $Y$)\n\n"
        "Paso 2:\n\n" +
        tb.tabulate(
            [
                ["Regresión", "$1$", f"$SCR = {SCR:.2f}$", f"${SCR:.2f}$", f"${F_c:.2f}$", f"${F_tab:.3f}$"],
                ["Residual (Error)", f"${n - 2}$", f"$SCE = {SCE:.2f}$", f"${SCR / F_c:.2f}$", "", ""],
                ["Total", f"${n - 1}$", f"$SCT = {SCT:.2f}$", "", "", ""],
            ],
            headers=[
                "Fuente de Variación",
                "GL",
                "SC",
                "CM",
                "$F_c$",
                "$F_{tab}$",
            ],
            tablefmt="github",
        ) +
        "\n\nPaso 3:\n"
    )
)

# Gráfica comparando F calculada y F tabulada
plt.figure(figsize=(10, 10))
plt.xlim(0, 30)
plt.ylim(-1, 1)
plt.axvline([F_tab], color="black", label=r"$F_{tab}$")
plt.axvline([F_c], color="red", label=r"$F_c$")
plt.show()

display(
    di.Markdown(
        "$$\n"
        rf"F_{{(1 - \alpha; \ 1; \ n - 2)}}"
        "\n$$\n\n"
        "$$\n"
        rf"F_{{\text{{tab}}}} = {F_tab:.3f} {'>' if F_c < F_tab else '<'} {F_c:.2f} = F_{{\text{{cal}}}}"
        "\n$$\n\n"
        "Paso 4:\n\n"
        "Conclusión:  \n"
        f"Con un nivel de significancia del {alpha:.2%} se {'rechaza' if F_c < F_tab else 'acepta'} la hipótesis nula, por lo que se concluye que {'no hay' if F_c < F_tab else 'si hay'} relación lineal entre $X$ y $Y$.\n\n"
    )
)
```

## Estimación de la varianza


Para estimar la varianza de los errores $\sigma^2$, podemos utilizar:

$$
\hat{\sigma}^2 = \frac{\sum_{i=1}^{n} e_i^2}{n}
$$

Que es estimador máximo verosímil de $\sigma^2$, pero es un estimador sesgado.

Un estimador insesgado de $\sigma^2$ es la varianza residual

$$
S_R^2 = \frac{\sum_{i=1}^{n} e_i^2}{n - 2}
$$

#### Ejercicio


Calcular la varianza residual con los adtos del ejercicio anterior.


```python
# Solución
recta_de_regresion = [
    [x[i], y[i], round(regresion(x[i]), 2), round((y[i] - regresion(x[i])), 2)] for i in range(n)
]

S_R = sum([dato[3] ** 2 for dato in recta_de_regresion]) / (n - 2)

display(
    di.Markdown(
        tb.tabulate(
            recta_de_regresion,
            headers=[
                r"$X$",
                r"$Y_i$",
                r"$\hat{Y}_i$",
                r"$e_i$",
            ],
            tablefmt="github"
        ) +
        "\n\n"
        "$$\n"
        r"S_R^2 = \frac{\sum_{i = 1}^n e_i^2}{n - 2} = "
        rf"\frac{{{sum([dato[3] ** 2 for dato in recta_de_regresion]):.2f}}}{{{n} - 2}} = {S_R:.2f}"
        "\n$$\n\n"
    )
)
```
