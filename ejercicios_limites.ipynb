```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
from IPython.display import Markdown, display

plt.style.use(
    "https://drive.usercontent.google.com/u/0/uc?id=11j2EGxQzAZ858aZhLr6duZPf_"
    "sR1zRkI&export=download"
)
```

```python
x, a = sp.symbols("x, a")
```

## Ejercicios

```python
a = (sp.root(2 * x + 4, 3) - 2) / (x**3 - 8)  # type: ignore
b = (sp.root(2 * x + 4, 3) - sp.root(2 * x - 1, 3)) / (x**2 - 3 * x + 2)  # type: ignore
c = (sp.sqrt(x + 1) + sp.root(2 * x + 2, 3) - 2) / (x**2 - 9)  # type: ignore
d = (x * sp.sqrt(x) - a * sp.sqrt(a)) / (sp.sqrt(x) - sp.sqrt(a))  # type: ignore
e = (sp.root(2 * x - 1, 3) + sp.root(x + 7, 3) - 3) / (x**2 - 3 * x + 2)  # type: ignore
f = (x * sp.sqrt(x + 2) + x**2 - 8) / (sp.sqrt(x + 2) - 2)  # type: ignore
g = (x**2 + sp.sqrt(x + 3) + x - 4) / (x**2 - x)  # type: ignore
h = (x**2 + 5 * sp.sqrt(2 * x + 5) - 6 * x - 7) / (x**4 - 16)  # type: ignore
ejercicios = [a, b, c, e, f, g, h]

display(
    Markdown(
        f"""$$
\\text{{a) }} \\lim_{{x \\to 2}} {sp.latex(a)} = {sp.latex(sp.limit(a, x, 2))}
$$
$$
\\text{{b) }} \\lim_{{x \\to 2}} {sp.latex(b)} = {sp.latex(sp.limit(b, x, 2))}
$$
$$
\\text{{c) }} \\lim_{{x \\to 3}} {sp.latex(c)} = {sp.latex(sp.limit(c, x, 3))}
$$
$$
\\text{{e) }} \\lim_{{x \\to 1}} {sp.latex(e)} = {sp.latex(sp.limit(e, x, 1))}
$$
$$
\\text{{f) }} \\lim_{{x \\to 2}} {sp.latex(f)} = {sp.latex(sp.limit(f, x, 2))}
$$
$$
\\text{{g) }} \\lim_{{x \\to 1}} {sp.latex(g)} = {sp.latex(sp.limit(g, x, 1))}
$$
$$
\\text{{h) }} \\lim_{{x \\to 2}} {sp.latex(h)} = {sp.latex(sp.limit(h, x, 2))}
$$"""
    )
)
```

```python
ejercicios = [a, b, c, e, f, g, h]
limites = [
    2,
    2,
    3,
    1,
    2,
    1,
    2,
]
respuestas = [
    sp.limit(ejercicio, x, limite) for ejercicio, limite in zip(ejercicios, limites)
]
nombres_ejercicios = ["a", "b", "c", "e", "f", "g", "h"]

x_ = np.linspace(0, 4, 1001)
y_ejercicios = []
x_ejercicios = []

for ejercicio in ejercicios:
    y_ = []
    x_ejercicio = []

    for x__ in x_:
        valor = sp.limit(ejercicio, x, x__)

        if valor.is_real:
            x_ejercicio.append(x__)
            y_.append(valor)

    x_ejercicios.append(x_ejercicio)
    y_ejercicios.append(y_)


fig, ax = plt.subplots(4, 2, figsize=(15, 15))

fig.subplots_adjust(hspace=0.25, wspace=0.25)

for i, ax_ in enumerate(ax.flatten()):
    if i < len(ejercicios):
        ax_.plot(
            x_ejercicios[i],
            y_ejercicios[i],
            label=f"${nombres_ejercicios[i]}(x) = {sp.latex(ejercicios[i])}$",
        )
        ax_.set_title(f"Ejercicio {nombres_ejercicios[i]}")
        ax_.axvline(limites[i], color="white", lw=1, label=f"$x = {limites[i]}$")

        if respuestas[i].is_real:
            ax_.axhline(
                respuestas[i],
                color="green",
                lw=1,
                label=f"$y = {sp.latex(respuestas[i])}$",
            )

        ax_.legend()
        ax_.set_xlabel("$x$")
        ax_.set_ylabel(f"${nombres_ejercicios[i]}(x)$")

plt.tight_layout()
plt.ylim(-2, 27)
plt.show()
```
